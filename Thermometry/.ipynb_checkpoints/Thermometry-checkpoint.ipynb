```python
import numpy as np                  
import matplotlib.pyplot as plt                     
import scipy as scp  
from scipy.optimize import curve_fit
import statistics as st
import sympy as sp 
from IPython.display import Latex
```

```python
m, a, ts, tc, pt = np.loadtxt('../Thermometry/Part A Data.txt', unpack=True)
t1, R1 = np.loadtxt('../Thermometry/400 ml Part C.txt', unpack=True)
t2, R2 = np.loadtxt('../Thermometry/1000 ml Part C.txt', unpack=True)
t3, R3 = np.loadtxt('../Thermometry/2000 ml Part C.txt', unpack=True)
t_h, R_h = np.loadtxt('../Thermometry/Heating Wax Part B.txt', unpack=True)
t_c, R_c = np.loadtxt('../Thermometry/Cooling Wax Part B.txt', unpack=True)
T, t = np.loadtxt('../Thermometry/Lees Disc Part E.txt', unpack=True)
```

```python
#Part A - Calibration
#Alcohol Thermometer
p1, q1 = np.polyfit(m[10:], a[10:], 1)

plt.plot(m, p1*m+q1, color='red')

print(p1, q1)

plt.scatter(m, a)
plt.xlabel(r'Mercury Theromometer Temperature [$^o$C]')
plt.ylabel(r'Alcohol Theromometer Temperature [$^o$C]')
plt.show()
```

```python
#Part A - Calibration
#Thermocouple
p2, q2 = np.polyfit(m[10:], -tc[10:], 1)

plt.plot(m, p2*m+q2, color='red')

print(p2)

plt.scatter(m, -tc)
plt.xlabel(r'Mercury Theromometer Temperature [$^o$C]')
plt.ylabel(r'Thermocouple Potential [V]')
plt.show()
```

```python
#Part A - Calibration
#Pt-100
p3, q3 = np.polyfit(m[10:], pt[10:], 1)

plt.plot(m, p3*m+q3, color='red')

print(p3, q3)

plt.scatter(m, pt)
plt.xlabel(r'Mercury Theromometer Temperature [$^o$C]')
plt.ylabel(r'Pt-100 Resistance [$\Omega$]')
plt.show()
```

```python
#Part A - Calibration
#Thermistor
def exp(x, a, b, c):
    return a+b*np.log(x)+c*(np.log(x))**3

par, cov = curve_fit(exp, ts[10:], m[10:])

R = np.linspace(10, 350, 3500)

plt.plot(R, par[0]+par[1]*np.log(R)+par[2]*(np.log(R))**3, color='red')

print(par[0], par[1], par[2])

plt.scatter(ts, m)
plt.ylabel(r'Mercury Theromometer Temperature [$^o$C]')
plt.xlabel(r'Thermistor Resistance [$\Omega$]')
plt.show()
```

```python
#Part C & D - Newton's Law of Cooling
T1 = (R1-103)/0.3065
T2 = (R2-103)/0.3065
T3 = (R3-103)/0.3065

T0 = 25

def cool1(x, a, b):
    return np.exp(a*x+np.log(T1[0]-T0))+b

def cool2(x, a, b):
    return np.exp(a*x+np.log(T2[0]-T0))+b

def cool3(x, a, b):
    return np.exp(a*x+np.log(T3[0]-T0))+b

P1, C1 = curve_fit(cool1, t1*30, T1, p0=[-0.01, T0])
P2, C2 = curve_fit(cool2, t2*30, T2, p0=[-0.01, T0])
P3, C3 = curve_fit(cool3, t3*30, T3, p0=[-0.01, T0])

R = np.linspace(0, 2500, 2500)

plt.plot(R, np.exp(P1[0]*R+np.log(T1[0]-T0))+P1[1], color='red')
plt.plot(R, np.exp(P2[0]*R+np.log(T2[0]-T0))+P2[1], color='red')
plt.plot(R, np.exp(P3[0]*R+np.log(T3[0]-T0))+P3[1], color='red')

plt.scatter(t1*30, T1, s=10, label='400 ml')
plt.scatter(t2*30, T2, s=10, label='1000 ml')
plt.scatter(t3*30, T3, s=10, label='2000 ml')
plt.xlabel(r'Time [s]')
plt.ylabel(r'Temperature [$^o$C]')
plt.legend()
plt.show()
```

```python
d = np.array([0.08266, 0.10558, 0.13202])
V = np.array([400, 1000, 2000])*10**(-6)
h = V/(np.pi*(d/2)**2)

k = np.array([P1[0], P2[0], P3[0]])

A = np.pi*(d/2)**2

H = (-k*997*100*4200*10**(-6))/A

print(H)

plt.scatter(A[0], H[0], label='400 ml')
plt.scatter(A[1], H[1], label='1000 ml')
plt.scatter(A[2], H[2], label='2000 ml')
plt.xlabel(r'Surface Area [$m^2$]')
plt.ylabel(r'Heat Coefficient [$W/m^2K$]')
plt.legend()
plt.show()
```

```python
#Part E - Lee's Disc

T_0 = 52.4
T_s = 25

def cool(x, a, b):
    return np.exp(a*x+np.log(T_0-T_s))+b

P, C = curve_fit(cool, t, T, p0=[-0.01, T_s])

t_lin = np.linspace(0, 1250, 12500)

print(P[0], P[1])

# plt.plot(t_lin, -0.008*(t_lin-415)+47.4)

plt.plot(t_lin, np.exp(P[0]*t_lin+np.log(T_0-T_s))+P[1], color='red')
plt.scatter(t, T)
plt.xlabel(r'Time [s]')
plt.ylabel(r'Temperature [$^o$C]')
plt.hlines(47.4, 0, 1254, ls='--', color='green', label=r'Steady State Temperature (47.4$^o$C)')
plt.legend()
plt.show()
```
