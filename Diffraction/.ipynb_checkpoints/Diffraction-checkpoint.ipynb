```python
# Surface Tension

L = 2.775
H = 0.138
lam = 6.3*10**(-7)

f = np.array([100, 110, 120, 130, 140, 150, 160, 170, 180, 190, 200])
x1 = np.array([0.8, 0.9 , 1.1, 1.3, 1.4, 1.4, 1.5, 1.7, 1.8, 1.8, 2])*10**(-2)/2+0.006

theta = np.arctan(H/L)

gamma = np.arctan((H+x1)/L)-theta

k = (2*np.pi/lam)*np.sin(gamma)*np.sin(theta)

print(np.round(k, 2))

omega = 2*np.pi*f

p, q = np.polyfit(np.log(k), np.log(omega), 1)

print(p, q)

plt.plot(np.log(k), p*np.log(k)+q, color='red')
plt.scatter(np.log(k), np.log(omega))
plt.xlabel(r'$\ln(k)$')
plt.ylabel(r'$\ln(\omega)$')
plt.show()
```
